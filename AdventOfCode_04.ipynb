```python
def parseInputData(data):
    lines = data.splitlines()
    
    numbers = [int(i) for i in lines[0].split(',')]
    
    boards = {}
    currentBoard = 0
    currentRow = 0
    board = [None]*5
    
    for line in lines[2:]:
        if line == '':
            boards[currentBoard] = board;
            currentBoard = currentBoard + 1
            currentRow = 0
            board = [None]*5
        else:
            board[currentRow] = [int(i) for i in line.split()]
            currentRow = currentRow + 1
    boards[currentBoard] = board;
    
    return (boards, numbers)
```

```python
def runBingoGame(data):
    boards,numbers = parseInputData(data)
    winningBoard,winningNumber = playBingo(boards, numbers)
    print(winningBoard, calculateScore(boards[winningBoard],winningNumber))
```

```python
def playBingo(boards, numbers):
    winner = -1
    number = -1
    while winner < 0:
        number = number + 1
        boards = callNumber(boards,numbers[number])
        winner = findWinner(boards)
    return winner, numbers[number]
```

```python
def calculateScore(board,lastNumber):
    boardSum = 0
    for i in range(5):
        for j in range(5):
            if board[i][j] > 0:
                boardSum = boardSum + board[i][j]
    return boardSum*lastNumber
```

```python
def callNumber(boards, calledNumber):
    for board in boards:
        for r, row in enumerate(boards[board]):
            for square, number in enumerate(row):
                if number == calledNumber:
                    boards[board][r][square] = -1
    return boards
```

```python
def findWinner(boards):
    for board in boards:
        for row in boards[board]:
            if row == [-1,-1,-1,-1,-1]:
                return board
        for x in range(5):
            isWin = True
            for y in range(5):
                if boards[board][y][x] != -1:
                    isWin = False
                    break
            if(isWin):
                return board;
    return -1
```

```python
def runAllBingoGames(data):
    boards,numbers = parseInputData(data)
    while len(boards) >= 1:
        winningBoard,winningNumber = playBingo(boards, numbers)
        print(winningBoard, calculateScore(boards[winningBoard],winningNumber))
        boards.pop(winningBoard)
```
