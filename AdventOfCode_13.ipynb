## Parse data

```python
from timeit import default_timer as timer

def parseData(data):
    start = timer()
    dots = {}
    instructions = []
    for line in data.splitlines():
        if line is not '':
            if line.startswith('fold along'):
                instructions.append(parseInstruction(line))
            else:
                dots = addDot(parseDot(line), dots)
    end = timer()
    print("parse time: "+"{:10.7f}".format(end-start))
    return dots, instructions

def parseInstruction(line):
    line = line.replace('fold along ', '')
    terms = line.split('=')
    return (terms[0], int(terms[1]))
    
def parseDot(line):
    coord = line.split(',')
    return [int(scalar) for scalar in coord]

def addDot(coord, dots):
    if(coord[1] not in dots.keys()):
        dots[coord[1]] = {}
    dots[coord[1]][coord[0]] = '#'
    return dots
```

## Utilities

```python
def printDots(dots):
    Y,X = getSheetSize(dots)
    for y in range(Y):
        for x in range(X):
            if y in dots.keys() and x in dots[y].keys():
                print('#', end = '')
            else:
                print('.', end = '')
        print()

def getSheetSize(dots):
    y = max(dots.keys())+1
    x = max([max(dots[row].keys()) for row in dots])+1
    return (y,x)
```

## Part 1

```python
def doOneFold(data):
    dots, instructions = parseData(data)
    start = timer()
    dots = applyFold(dots, instructions[0])
    totalDots = countDots(dots)
    end = timer()
    print("run time: "+"{:10.7f}".format(end-start))
    print(totalDots)

def applyFold(dots, instruction):
    if instruction[0] is 'y':
        #print("folding on "+str(instruction[1]))
        for n in range(1, getSheetSize(dots)[0]-instruction[1]):
            #print("merging "+str(instruction[1]+n)+" and "+str(instruction[1]-n))
            if instruction[1]+n in dots.keys():
                if instruction[1]-n not in dots.keys():
                    dots[instruction[1]-n] = {**dots[instruction[1]+n]}
                else:
                    dots[instruction[1]-n] = {**dots[instruction[1]-n], **dots[instruction[1]+n]}
                del dots[instruction[1]+n]
        if instruction[1] in dots.keys():
            del dots[instruction[1]]
    else:
        for y in dots:
            for n in range(1, getSheetSize(dots)[1]-instruction[1]):
                if instruction[1]+n in dots[y].keys():
                    if instruction[1]-n not in dots[y].keys():
                        dots[y][instruction[1]-n] = dots[y][instruction[1]+n]
                    del dots[y][instruction[1]+n]
            if instruction[1] in dots[y].keys():
                del dots[y][instruction[1]]
    return dots

def countDots(dots):
    return sum(len(dots[line]) for line in dots)
```

## Part 2

```python
def doAllFolds(data):
    dots, instructions = parseData(data)
    start = timer()
    for instruction in instructions:
        dots = applyFold(dots, instruction)
    end = timer()
    print("run time: "+"{:10.7f}".format(end-start))
    printDots(dots)
```
