## Parse data

```python
from timeit import default_timer as timer

def parseData(data):
    start = timer()
    parsedData = [parseLine(line) for line in data.splitlines()]
    end = timer()
    print("parse time: "+"{:10.7f}".format(end-start))
    return parsedData

def parseLine(line):
    return [int(location) for location in line]
```

## Part 1 - Find the low points

```python
from collections import namedtuple
Point = namedtuple('Point', ['x','y'])

def calculateRiskLevel(data):
    map = parseData(data)
    start = timer()
    lowPoints = findLowPoints(map)
    totalRiskLevel = calculateTotalRiskLevel(map,lowPoints)
    end = timer()
    print("run time: "+"{:10.7f}".format(end-start))
    print(totalRiskLevel)
    
def findLowPoints(map):
    lowPoints = []
    for x in range(len(map)):
        for y in range(len(map[x])):
            if isLowPoint(map, x, y):
                lowPoints.append(Point(x,y))
    return lowPoints

def isLowPoint(map, x, y):
    if ((x-1 >= 0) and (map[x][y] >= map[x-1][y])
    or (x+1 < len(map)) and (map[x][y] >= map[x+1][y])
    or (y-1 >= 0) and (map[x][y] >= map[x][y-1])
    or (y+1 < len(map[x])) and (map[x][y] >= map[x][y+1])):
        return False
    return True

def calculateTotalRiskLevel(map, points):
    total = 0
    for point in points:
        total += map[point.x][point.y]+1
    return total
```

## Part 2 - Calculate basin sizes

```python
from functools import reduce

def calculateLargestBasinSizes(data):
    map = parseData(data)
    start = timer()
    lowPoints = findLowPoints(map)
    basinSizes = [mapBasinSize(lowPoint, map) for lowPoint in lowPoints]
    basinSizes.sort(reverse=True)
    largestBasins = reduce((lambda x,y: x*y), basinSizes[:3])
    end = timer()
    print("run time: "+"{:10.7f}".format(end-start))
    print(largestBasins)
    
def mapBasinSize(point, map):
    basinPoints = [point]
    searchPoints = [point]
    while len(searchPoints) != 0:
        newSearchPoints = []
        for p in searchPoints:
            newNeighbours = findNon9Neighbours(p, map)
            newSearchPoints += [neighbour for neighbour in newNeighbours if (neighbour not in newSearchPoints) and (neighbour not in basinPoints)]
        basinPoints += newSearchPoints
        searchPoints = newSearchPoints
    return len(basinPoints)

def findNon9Neighbours(point, map):
    neighbours = []
    if (point.x-1 >= 0) and (map[point.x-1][point.y] < 9):
        neighbours.append(Point(point.x-1,point.y))
    if (point.x+1 < len(map)) and (map[point.x+1][point.y] < 9):
        neighbours.append(Point(point.x+1,point.y))
    if (point.y-1 >= 0) and (map[point.x][point.y-1] < 9):
        neighbours.append(Point(point.x,point.y-1))
    if (point.y+1 < len(map[point.x])) and (map[point.x][point.y+1] < 9):
        neighbours.append(Point(point.x,point.y+1))
    return neighbours
```
