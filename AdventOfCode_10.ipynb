## Parse Data 

```python
opening = ['(','[','{','<']
closing = [')',']','}','>']

class Fragment:
    close = ''
    
    def __init__(self, open):
        self.open = open
        self.children = []
    
    def __str__(self):
        return self.open+''.join([child.__str__() for child in self.children])+self.close
    
    def __repr__(self):
        return self.__str__()
    
    def isThisComplete(self):
        return self.open in opening and self.close in closing
    
    def isComplete(self):
        return (self.isThisComplete() and all([child.isComplete() for child in self.children]))
    
    def isThisCorrupt(self):
        return (self.open+self.close) not in ['()','[]', '{}','<>','(','[', '{','<']
    
    def isCorrupt(self):
        return (self.isThisCorrupt() or any([child.isCorrupt() for child in self.children]))
    
    def addChild(self, child):
        self.children.append(child)
```

```python
def parseData(data):
    return [parseLine(line) for line in data.splitlines()]

def parseLine(line):
    frags = []
    while line is not None:
        frag, line = parseFragment(line)
        frags.append(frag)
    return frags

def parseFragment(line):
    head, tail = popHead(line)
    frag = Fragment(head)
    
    while tail is not None and tail[0] in opening:
        childFrag, tail = parseFragment(tail)
        frag.children.append(childFrag)
    
    if tail is None:
        return frag, None
    
    if tail[0] in closing:
        head, tail = popHead(tail)
        frag.close = head
        return frag, tail
    
def popHead(line):
    if len(line) == 1:
        return line, None
    return line[0], line[1:]
```

## Part 1

```python
def scoreCorruptLines(data):
    lines = [line for line in parseData(data) if any(fragment.isCorrupt() for fragment in line)]
    illegalCharacters = [findFirstIllegalCharacter(line) for line in lines]
    print(scoreCharacters(illegalCharacters))
    
def findFirstIllegalCharacter(frags):
    root = next(frag for frag in frags if frag.isCorrupt())
    if(root.isThisCorrupt()):
        return root.close
    return findFirstIllegalCharacter(root.children)
    
def scoreCharacters(characters):
    return sum([scoreCharacter(character) for character in characters])

def scoreCharacter(character):
    if character == ')':
        return 3
    if character == ']':
        return 57
    if character == '}':
        return 1197
    if character == '>':
        return 25137
```

## Part 2

```python
import math

closingMap = {k:v for k, v in zip(opening, closing)}
completionScore = {')':1,']':2,'}':3,'>':4}

def fixIncompleteLines(data):
    lines = [line for line in parseData(data) if any(not fragment.isComplete() for fragment in line)]
    incompleteStrings = [lineToString(line) for line in lines]
    for line in lines:
        for frag in line:
            completeFrag(frag)
    completeStrings = [lineToString(line) for line in lines]
    newChars = [new.replace(old,'') for old, new in zip(incompleteStrings, completeStrings)]
    scores = [calculateScore(line) for line in newChars]
    scores.sort()
    print(scores[math.floor(len(scores)/2)])
    
def lineToString(line):
    return ''.join([str(frag) for frag in line])

def completeFrag(frag):
    if not frag.isThisComplete():
        frag.close = closingMap[frag.open]
    if frag.isComplete():
        return
    [completeFrag(child) for child in frag.children]
    
def calculateScore(line):
    total = 0;
    for char in line:
        total *= 5
        total += completionScore[char]
    return total
```
