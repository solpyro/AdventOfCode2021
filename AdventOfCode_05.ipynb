```python
def parseData(data):
    return [parseLine(line) for line in data.splitlines()]
```

```python
def parseLine(line):
    start,end = line.split(' -> ')
    x1,y1 = start.split(',')
    x2,y2 = end.split(',')
    return(int(x1),int(y1),int(x2),int(y2))
```

```python
def mapStraightVents(data):
    lines = [line for line in parseData(data) if isStraight(line)]
    ventMap = populateStraightLines(lines)
    printMap(ventMap) 
    print(countOverlaps(ventMap))
```

```python
def isStraight(line):
    return (line[0] == line[2]) or (line[1] == line[3])
```

```python
def populateStraightLines(lines):
    ventMap = {}
    
    for line in lines:
        ventMap = addStraightLine(line, ventMap)
    
    return ventMap
```

```python
def countOverlaps(ventMap):
    count = 0
    for x in ventMap:
        for y in ventMap[x]:
            if ventMap[x][y] >= 2:
                count = count + 1
    return count
```

```python
def addStraightLine(line, ventMap):
    for x in range(min(line[0],line[2]),max(line[0],line[2])+1):
        for y in range(min(line[1],line[3]),max(line[1],line[3])+1):
            ventMap = incrementCell(x,y,ventMap)
            
    return ventMap
```

```python
def incrementCell(x,y,ventMap):
    if x not in ventMap:
        ventMap[x] = {} 
        
    if y not in ventMap[x]:
        ventMap[x][y] = 0
        
    ventMap[x][y] = ventMap[x][y]+1
    
    return ventMap
```

```python
def printMap(ventMap):
    w = max(ventMap.keys())+1
    h = 0
    for x in ventMap:
        h = max(h, max(ventMap[x].keys())+1)
    print(w,h)
    for y in range(h):
        for x in range(w):
            if (x not in ventMap) or (y not in ventMap[x]):
                print('.',end='')
            else:
                print(ventMap[x][y],end='')
        print()
```

```python
def mapAllVents(data):
    lines = parseData(data)
    ventMap = populateAllLines(lines)
    #printMap(ventMap) 
    print(countOverlaps(ventMap))
```

```python
def populateAllLines(lines):
    ventMap = {}
    
    for line in lines:
        if(isStraight(line)):
            ventMap = addStraightLine(line, ventMap)
        else:
            ventMap = addDiagonalLine(line, ventMap)
    
    return ventMap
```

```python
def addDiagonalLine(line, ventMap):
    for x, y in zip(directionalRange(line[0],line[2]), directionalRange(line[1],line[3])):
        ventMap = incrementCell(x,y,ventMap)
            
    return ventMap
```

```python
def directionalRange(a,b):
    if b >= a:
        return range(a,b+1)
    return range(a, b-1, -1)
```
