## Part 1

```python
from timeit import default_timer as timer

def parseCommands_v1(commands):
    start = timer()
    position = (0,0)
    for command in commands:
        position = parseCommand_v1(command, position)
    end = timer()
    print("time: "+"{:10.7f}".format(end-start))
    print(position, position[0]*position[1])

def parseCommand_v1(command, position):
    direction,distance = command.split(' ')
    distance = int(distance)
    if direction == 'forward':
        return (position[0]+distance,position[1])
    elif direction == 'down':
        return (position[0],position[1]+distance)
    elif direction =='up':
        return (position[0],position[1]-distance)
```

```python
def parseCommands_v2(commands):
    start = timer()
    position = (0,0,0)
    for command in commands:
        position = parseCommand_v2(command, position)
    end = timer()
    print("time: "+"{:10.7f}".format(end-start))
    print(position, position[0]*position[1])

def parseCommand_v2(command, position):
    direction,distance = command.split(' ')
    distance = int(distance)
    if direction == 'forward':
        return (position[0]+distance,position[1]+(position[2]*distance),position[2])
    elif direction == 'down':
        return (position[0],position[1],position[2]+distance)
    elif direction =='up':
        return (position[0],position[1],position[2]-distance)
```

```python
parseCommands_v2(testData)
```
